```python
#Joining DataFrames Using .join()
data1 = {'Name':['Jai', 'Princi', 'Gaurav', 'Anuj'],
        'Age':[27, 24, 22, 32]}

data2 = {'Address':['Allahabad', 'Kannuaj', 'Allahabad', 'Kannuaj'],
        'Qualification':['MCA', 'Phd', 'Bcom', 'B.hons']}

df1 = pd.DataFrame(data1,index=['K0', 'K1', 'K2', 'K3'])

df2 = pd.DataFrame(data2, index=['K0', 'K2', 'K3', 'K4'])

df3=df1.join(df2)
df3
```

```python
#Now we use how = 'outer' in order to get union
df3=df1.join(df2,how='outer')
df3
```

```python
#Joining DataFrames Using the "on" Argument
data1 = {'Name':['Jai', 'Princi', 'Gaurav', 'Anuj'],
        'Age':[27, 24, 22, 32],
        'Key':['K0', 'K1', 'K2', 'K3']}

data2 = {'Address':['Allahabad', 'Kannuaj', 'Allahabad', 'Kannuaj'],
        'Qualification':['MCA', 'Phd', 'Bcom', 'B.hons']}

df1=pd.DataFrame(data1)
df2=pd.DataFrame(data2,index=['K0', 'K2', 'K3', 'K4'])
print(df1, "\n\n", df2)
```

```python
df3=df1.join(df2,on='Key')
df3
```

```python
import pandas as pd

df = pd.DataFrame({
    "employee_id": [1,2,3,4,5,6,7,8],
    "employee": ["A","B","C","D","E","F","G","H"],
    "department": ["IT","HR","IT","HR","IT","Sales","Sales","IT"],
    "salary": [100, 80, 120, 90, 130, 150, 140, 110],
    "bonus": [10, 5, 12, 7, 15, 20, 18, 10]
})
df
```

```python
# Total employees per department
df.groupby('department')['employee_id'].count()
```

```python
#Average salary per department
# ------  select department, avg(salary) from employee group by department
df.groupby('department')['salary'].mean()
```

```python
# Departments with more than 2 employees (HAVING)
df.groupby('department').filter(lambda x : len(x) >2 )
```

```python
#  Add each department's average salary to each row (window function)
df['avg_dept']=df.groupby('department')['salary'].transform("mean")
```

```python
df
```

```python
#  Highest salary in each department
df.groupby('department')['salary'].max()
```

```python
# Total salary + bonus per department
'''
SELECT department,
       SUM(salary) AS total_salary,
       SUM(bonus) AS total_bonus
FROM employees
GROUP BY department;

'''
df.groupby('department').agg(
    total_salary=('salary','sum'),
    total_bonus=('bonus','sum')
)
```

```python
# Find employees whose salary is above their departmentâ€™s average
df["dept_avg"]=df.groupby('department')['salary'].transform("mean")
```

```python
df
```

```python
df[df['salary']>df['dept_avg']]
```

```python
# Percentage contribution of each employee to their department salary total
df['sal_contribution']=df.groupby('department')['salary'].transform('sum')
```

```python
df
```

```python
df['pct']=df['salary']/df['sal_contribution']
```

```python
df
```

```python
# Salary rank within each department
df['sal_rank']=df.groupby('department')['salary'].rank(ascending=False)
```

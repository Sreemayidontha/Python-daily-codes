```python

## Explore function scope and variable accessibility. 
#Local Scope
def local_scope_example():
    x = 10  # Local variable
    print(x)  # Accessible inside the function

# Trying to access x outside the function will raise an error
# print(x)  # Uncommenting this will result in a NameError because x is not accessible outside

```

```python
#Global Scope
x = 5  # Global variable

def global_scope_example():
    print(x)  # Can access the global variable

global_scope_example()  # Output: 5

# The global variable is still accessible outside the function
print(x)  # Output: 5
```

```python
# Local and Global Variable with Same Name
x = 10  # Global variable

def scope_example():
    x = 5  # Local variable, shadows global x
    print(f"Local x: {x}")  # Output will be 5

scope_example()
print(f"Global x: {x}")  # Output will be 10, because global x is not affected
```

```python
## Write recursive functions.
###[function calls itself]

#factorial
def factorial(n):
  if n==0:
    return 1
  else:   
      return n*factorial(n-1)

n=int(input())
print(f"The factorial of {n} is {factorial(n)}")
```

```python
def fibanacci(n):
    if n==0:
        return 0
    elif n==1:
        return 1
    else:
        return fibanacci(n-1)+fibanacci(n-2)


n=6
print(f"The fibanacci of {n} is {fibanacci(n)}")
```

```python
def sum_of_digits(n):
    if n==0:
        return 0
    else:
        return n%10+sum_of_digits(n//10)
n=12345
print(f"The sum of digits of {n} is {sum_of_digits(n)}")
```

```python
#data cleaning and manipulation tasks.

# Display the first few rows to understand the data
df3.head()

# Check for missing values
missing_values = df3.isnull().sum()

# Check data types of each column
data_types = df3.dtypes

# Check for duplicates
duplicate_rows = df3.duplicated().sum()

# Summary statistics
summary_stats = df3.describe()

missing_values, data_types, duplicate_rows, summary_stats
```

```python
# data analysis and visualization using Pandas.

import matplotlib.pyplot as plt
import seaborn as sns

# Convert Date column to datetime
df3['Date'] = pd.to_datetime(df3['Date'])

# 1. Sales over Time (Total Amount vs Date)
plt.figure(figsize=(10, 6))
df3.groupby('Date')['Total Amount'].sum().plot()
plt.title('Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales Amount')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()

# 2. Age Distribution
plt.figure(figsize=(10, 6))
sns.histplot(df3['Age'], bins=20, kde=True, color='skyblue')
plt.title('Age Distribution of Customers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# 3. Product Category Distribution
plt.figure(figsize=(10, 6))
sns.countplot(data=df3, x='Product Category', palette='Set2')
plt.title('Product Category Distribution')
plt.xlabel('Product Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# 4. Gender Distribution
plt.figure(figsize=(8, 8))
sns.countplot(data=df3, x='Gender', palette='Set1')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
#pivot tables and group data for analysis.

# 1. Total Sales by Product Category
pivot_category_sales = pd.pivot_table(df3, 
                                     values='Total Amount', 
                                     index='Product Category', 
                                     aggfunc='sum', 
                                     sort=False)
#2. Total Quantity Sold by Product Category
pivot_category_quantity = pd.pivot_table(df3, 
                                         values='Quantity', 
                                         index='Product Category', 
                                         aggfunc='sum', 
                                         sort=False)
pivot_category_sales,pivot_category_quantity
```

```python
# Creating a list
my_list = [1, 2, 3, 4, 5]

# Indexing and slicing
print(my_list[2])      # 3 (indexing)
print(my_list[:3])     # [1, 2, 3] (slicing)

# Adding and removing elements
my_list.append(6)      # Add 6 to the list
my_list.remove(3)      # Remove the element '3'
print(my_list)         # [1, 2, 4, 5, 6]
```

```python
# Creating a tuple
my_tuple = (1, 2, 3)

# Indexing (same as lists)
print(my_tuple[1])  # 2

# Tuples are immutable, so we can't modify them directly
```

```python
# Creating a set
my_set = {1, 2, 3, 4, 5}

# Adding and removing elements
my_set.add(6)    # Add 6
my_set.remove(3) # Remove 3
print(my_set)    # {1, 2, 4, 5, 6}

# Set operations: union, intersection
another_set = {4, 5, 6, 7}
print(my_set & another_set)  # Intersection: {4, 5, 6}
print(my_set | another_set)  # Union: {1, 2, 4, 5, 6, 7}
```

```python
#operator precedence.

# Define variables
a = 15
b = 3
c = 5

# Arithmetic operations
result = (a + b) * c  # Adding first, then multiplying
print(f"(a + b) * c = {result}")  # (15 + 3) * 5 = 90

# Combined operations
total = a - b + (c * 2)  # First multiplication, then addition and subtraction
print(f"a - b + (c * 2) = {total}")  # 15 - 3 + (5 * 2) = 22
```

```python
x = 10
y = 20
z = 10

#operators in expressions and calculations.
# Comparing values
print(x == z)  # True, since 10 == 10
print(x != y)  # True, since 10 != 20
print(x < y)   # True, since 10 < 20
print(x >= z)  # True, since 10 >= 10
```
